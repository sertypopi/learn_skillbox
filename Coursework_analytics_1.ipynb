```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
#импортирую модули и уточняю, что графики должны отражаться в данном файле

students = pd.DataFrame(pd.read_csv('students.csv', encoding='UTF-8'))
students = students.rename(columns={'id': 'student_id'})
#загружаю информацию о студентах в датафрейм

mean_stud = students[:]

#отдельный датафрейм для операций с возрастом

c = pd.DataFrame(pd.read_csv('courses.csv', encoding='UTF-8'))
c_c = pd.DataFrame(pd.read_csv('course_contents.csv', encoding='UTF-8'))
p = pd.DataFrame(pd.read_csv('progresses.csv', encoding='UTF-8'))
p_p = pd.DataFrame(pd.read_csv('progress_phases.csv', encoding='UTF-8'))                       
#загружаю информацию о курсах в датафреймы
 
c = c.rename(columns={'id': 'course_id'})

p = p.rename(columns={'id':'progress_id'})
#переименовываю столбцы в датафреймах курсов и прогресса

main_0 = pd.merge(p, c, on='course_id')
main1 = pd.merge(p_p, main_0, on='progress_id')
#объединяю дф прогресс с курсами, получившийся дф с фазами прогресса

main1.set_index(['course_id', 'module_number', 'lesson_number'], inplace=True)
c_c.set_index(['course_id', 'module_number', 'lesson_number'], inplace=True)
#объявляю мультиндекс для предыдущего дф и дф с контентом курсов

main = pd.merge(main1, c_c, left_index=True, right_index=True)
#объединяю дф по индексам

del main['Unnamed: 0']
#удаляю лишний столбец

main = main.reset_index()
#сбрасываю мультииндекс


```

```python
def func_dif(x):
    x = x.copy()
    
    x1 = list(x['student_id_s'][1:].copy())
    y = list(x['student_id_d'][:-1].copy())
    #принимаем значения стартового столбца срез с 1 индекса, значения итогового
    #столбца срез до последнего индекса, обращаем их в списки
    
    f = zip(x1, y)
    s = []
    #хешируем получившиеся списки, создаем пустой список
    
    for i, j in f:
        d = i / j
        s.append(d)
    #проходимся в цикле позначениям, получившегося словаря, добавляем в
    #пустой список
        
    x['dif'] = 0
    x['dif'][0] = 1
    x.loc[x['dif'] == 0, 'dif'] = s
    x['dif'][0] = 0
    #создаем столбец со значениями 0, делаем 0 индекс - 1, где индекс равен
    #0 объявлем значения списком,делаем 0 индекс - 0
    return x
#функция, где создаем столбец конверсии
```

```python
java_05 = func_dif(java_05)

webdesbas5 = func_dif(webdesbas5)

excel5 = func_dif(excel5)

dig_proj5 = func_dif(dig_proj5)

web_25 = func_dif(web_25)

js_05 = func_dif(js_05)

anim5 = func_dif(anim5)

webdespro5 = func_dif(webdespro5)

webdev5 = func_dif(webdev5)

UXdes5 = func_dif(UXdes5)

mark_ing5 = func_dif(mark_ing5)

weddes_05 = func_dif(weddes_05)

smm_mark5 = func_dif(smm_mark5)

php15 = func_dif(php15)

java5 = func_dif(java5)
#применям ф-ю к д/ф
```

Для улучшения качества контента полезно выявить причину.
```python
def mark(x):
    f = 0
    for k in x:
        if k == 'fail':
            f -= 1
        elif k == 'done':
            f += 2
        elif k == 'wait':
            f += 0.5
        elif k == 'start':
            f += 0.1
    return f    
#функция для вычисления успеваемости: если задание сдано +1, если провалено -1
#задание в ожидании +0.5, только начато +0.5, если студент получает 
#0 или меньше, то мы его отнесём к неуспевающим, т.к. либо ему 
#сложно и он больше провалил, чем сдал или начинал и не сдавал
```
